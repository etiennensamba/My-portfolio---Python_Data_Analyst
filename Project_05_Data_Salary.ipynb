In this project, we will try to analyze whether experience time interferes with salary income

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings

# Disabling warnings/Desabilitar avisos
warnings.filterwarnings('ignore')
```

```python
Project = pd.read_csv('Salary_Data.csv')
Project.head()
```

```python
Project.shape
```

```python
Project.isnull().sum()
```

```python
plt.figure(figsize=(5, 3))
sns.heatmap(Project.isnull())
plt.title('Checking null fields', fontsize=20, fontweight=600);
```

```python
Project.describe()
```

```python
plt.figure( figsize=(10, 5) )
plt.style.use('ggplot')
plt.title('Salary per experience', fontsize=20, fontweight=600)
plt.plot(Project['YearsExperience'], Project['Salary'], color='black')
plt.xlabel('Year')
plt.ylabel('Salary');
```

We realize that the salary is proportional to the years of experience. 
More experience, higher annual salary

```python
plt.subplots( figsize=(15, 4) )
plt.suptitle('DENSITY', fontsize=20, fontweight=600)

plt.subplot(1,2,1)
plt.title('Salary density', fontsize=10, fontweight=600)
sns.kdeplot( Project['Salary'], color='black')
plt.xlabel('Salary', color='black')
plt.ylabel('Density', color='black')

plt.subplot(1,2,2)
plt.title('Experience years density', fontsize=10, fontweight=600)
sns.kdeplot( Project['YearsExperience'], color='black')
plt.xlabel('Years Experience', color='black')
plt.ylabel('Density', color='black');
```

Most workers receive between 50.000US$ and 75.000US$ with a small group between 100.000 and 120.000US$ per year
Most workers have between 2 and 5 years of experience.

```python
plt.subplots( figsize=(15, 4) )
plt.suptitle('Salary distribution', fontsize=20, fontweight=600)

plt.subplot(1,2,1)
sns.scatterplot( data=Project, x='Salary', y='YearsExperience')
plt.ylabel('Years Experience', color='black')
plt.xlabel('Salary', color='black')


plt.subplot(1,2,2)
sns.regplot( data=Project, x='Salary', y='YearsExperience')
plt.ylabel('Years Experience', color='black')
plt.xlabel('Salary', color='black');

#plt.savefig('teste.png', format='png')
```

We realize that the salary is proportional to the years of experience. 
More experience, higher annual salary

```python
plt.subplots( figsize=(15, 4) )

plt.subplot(1,2,1)
plt.title('Years experience box distribuition', fontsize=15, fontweight=600)
sns.boxplot(Project['YearsExperience'])
plt.xlabel('Years Experience', color='black')

plt.subplot(1,2,2)
plt.title('Salary box distribution', fontsize=15, fontweight=600)
sns.boxplot(Project['Salary'])
plt.xlabel('Salary', color='black');
```

```python
Project.corr('pearson')
```

```python
plt.figure(figsize=(10,5))
plt.title('Correlation', fontsize=20, fontweight=600)
Correlation = Project.corr('pearson')
sns.heatmap(Correlation, annot=True, cbar=True);
```

We noticed a strong correlation of 98% between the two variables
