In this project we will analyze the evolution of Brazilian energy companies

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

import plotly.express as px
import plotly.graph_objects as Dash

import warnings
warnings.filterwarnings('ignore')
```

```python
Project = pd.read_excel('Dados_Empresas_Energia.xlsx')
Project.head()
```

```python
Project = Project.set_index('Data')
Project.head(2)
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16, 7))
plt.title('Analysis of shares of Brazilian energy companies', loc='left', fontsize=25, fontweight=600 )


plt.plot( Project.index, Project['Petrobras'], color='#008c4a', linewidth=4, alpha=0.7 )
plt.text( Project.index[-1], Project['Petrobras'].tail(1), 'Petrobras', color='#008c4a', size='large', 
         horizontalalignment='left' )

for every_columns in Project.columns[1:]:
    plt.plot(Project.index, Project[every_columns], color='gray', linewidth=2, alpha=0.7)
    plt.text(Project.index[-1], Project[every_columns].tail(1), 
             every_columns, color='blue', size='large', horizontalalignment='left' )

plt.xlabel('Period', fontsize=12, fontweight=600)
plt.ylabel('Closing Price (R$)', fontsize=12, fontweight=600);
```
